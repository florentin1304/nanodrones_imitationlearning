```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import time
from IPython.display import clear_output

# Call clear_output() to clear the output of the current cell

def show_image(image_path):
    # Read the image
    
    img = mpimg.imread(image_path)
    
    # Display the image
    plt.imshow(img)
    plt.axis('off')  # Hide axis
    plt.show()

for i in range(500,1000):
    image_path = f"/home/florentin/Documents/webots_simulations/bitcraze_rl/controllers/gate_following_controller/data/images/run_y24-m04-d07_h15-m34_mkbru_img{i}.png"  # Change this to your image file path
    print(i)
    show_image(image_path)
    time.sleep(1/32)
    clear_output(wait=True)
```
