```python
import torch
import torchinfo
from my_tcn import TemporalConvNet
from pytorch_tcn import TCN
import numpy as np
# Example usage:
num_inputs = 128
l = 10
num_channels = [128 for i in range(1 + int(np.log2(l)))]
kernel_size = 2
causal = True
input_format = 'BSF'  # or 'BFS'

my_tcn = TemporalConvNet(num_inputs, num_channels, kernel_size, causal, input_format)
tcn_lib = TCN(
     num_inputs=num_inputs,
    num_channels=num_channels,
    kernel_size = kernel_size,
    causal = True,
    use_norm = 'layer_norm',
    input_shape = 'NLC',
    use_skip_connections = False
)

x = torch.randn(1, l, num_inputs) 
torchinfo.summary(tcn_lib, input_data=(x), depth=10, 
                  col_names=[ "input_size",
                "output_size",
                "num_params",
                "params_percent",
                "kernel_size",
                "mult_adds"]
                )
```
